# Using the TimeSeries Class and the GEV Estimators

```python
import xtremes as xx
import numpy as np
import matplotlib.pyplot as plt
import xtremes.HigherOrderStatistics as hos
import xtremes.miscellaneous as misc
import xtremes.bootstrap as bst
```

```python
TS = hos.TimeSeries(n=1000,modelparams=(0.5,0.1,0.1))
TS.simulate(rep=200)
TS.get_blockmaxima(block_size=10)
```

```python
TS.plot(rep=[1,2])
```

```python
TS.plot_blockmaxima(rep=1)
TS.plot_blockmaxima(rep=1, plot_type='hist')
```

```python
PWM = hos.PWM_estimators(TS)
PWM.get_PWM_estimation()
PWM.get_statistics(gamma_true=0)
PWM.get_CIs()
PWM.statistics
```

```python
#PWM.get_CIs(method='minimal_width')
PWM.statistics
```

```python
PWM.plot()
```

```python
MLE = hos.ML_estimators(TS)
MLE.get_ML_estimation(PWM_estimators= PWM)
MLE.get_statistics(gamma_true=0)
MLE.get_CIs()  
MLE.plot()  
```

```python
MLE = hos.Frechet_ML_estimators(TS)
MLE.get_ML_estimation(PWM_estimators= PWM)
MLE.get_statistics(alpha_true=0)
MLE.get_CIs()  
MLE.plot()  
```
